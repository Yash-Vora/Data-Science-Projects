# Problem Statement : Build machine learning model that can predict price of the house based on different features like address,rooms,type,seller etc.

# Import required libraries for Linear Regression

```python
import pandas as pd  # For data analysis
import numpy as np  # For data analysis
import matplotlib.pyplot as plt  # For data visualization
%matplotlib inline
import seaborn as sns  # For data visualization
from sklearn.linear_model import LinearRegression # For regression in machine learning
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error  # For accuracy,mean squared error,mean absolute error
from sklearn.model_selection import train_test_split  # For separating training and testing data
```

# Data Exploration and Pre Processing

Load the given dataset

```python
homeprices = pd.read_csv('homeprices.csv')
homeprices
```

Print all the column names

```python
homeprices.columns
```

Describe the data

```python
homeprices.describe()
```

Drop address, date, postcode, YearBuilt, lattitude, longtitude columns

```python
homeprices = homeprices.drop(['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'],axis=1)
homeprices
```

Find the count of null value in each column

```python
homeprices.isnull().sum()
```

Fill the null value of property count, distance, Bedroom2, Bathroom, Car with 0

```python
homeprices[['Propertycount','Distance','Bedroom2','Bathroom','Car']] = homeprices[['Propertycount','Distance','Bedroom2','Bathroom','Car']].fillna(0)
homeprices[['Propertycount','Distance','Bedroom2','Bathroom','Car']]
```

Fill Null value of land size and bidding area columns with Mean

```python
homeprices[['Landsize','BuildingArea']] = homeprices[['Landsize','BuildingArea']].fillna(homeprices[['Landsize','BuildingArea']].mean())
homeprices[['Landsize','BuildingArea']]
```

```python
# Drop whole row where there is null values in price column
homeprices.dropna(inplace=True)
```

Find the unique value in method column

```python
homeprices['Method'].unique()
```

Create a dummy data for categorical data

```python
# Converting categorical data into numerical data using one hot encoding
homeprices = pd.get_dummies(homeprices,drop_first=True)
homeprices
```

# Building the model and check the accuracy

Create the target data and feature data where target data is price

```python
# Separating predictor columns
x = homeprices.drop(['Price'],axis=1)
x
```

```python
# Separating target column
y = homeprices[['Price']]
y
```

```python
# Split training and testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

Create a linear regression model for Target and feature data

```python
# Building the linear regression model using training data
linear_regression = LinearRegression()
linear_regression.fit(x_train,y_train)
```

Check if the model is overfitting or underfitting or it is accurate

```python
# Accuracy for testing data set
linear_regression.score(x_test,y_test)
```

```python
# Accuracy for training data set
linear_regression.score(x_train,y_train)
```

If the model is overfitting then apply ridge and lasso regression algorithms

```python
from sklearn.linear_model import Ridge,Lasso
```

# Ridge Regression

```python
# Building the ridge regression model using training data
ridge_regression = Ridge(alpha=50,max_iter=100)
ridge_regression.fit(x_train,y_train)
```

```python
# Accuracy for testing data set
ridge_regression.score(x_test,y_test)
```

```python
# Accuracy for training data set
ridge_regression.score(x_train,y_train)
```

# Lasso Regression

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Building the lasso regression model using training data
lasso_regression = Lasso(alpha=50,max_iter=100)
lasso_regression.fit(x_train,y_train)
```

```python
# Accuracy for testing data set
lasso_regression.score(x_test,y_test)
```

```python
# Accuracy for training data set
lasso_regression.score(x_train,y_train)
```

Extract slope and intercept value from the model

```python
print("Slope :",lasso_regression.coef_)
```

```python
print("\nIntercept :",lasso_regression.intercept_)
```

Display Mean Squared Error

```python
predicted_value = lasso_regression.predict(x_test)
predicted_value
```

```python
y_test
```

```python
# Mean Squared Error
mean_squared_error(y_test,predicted_value)
```

Display Mean Absolute Error

```python
# Mean Absolute Error
mean_absolute_error(y_test,predicted_value)
```

Display Root mean Squared error

```python
# Root Mean Squared Error
np.sqrt(mean_squared_error(y_test,predicted_value))
# or
# mean_squared_error(y_test,predicted_value) ** (1/2)
```

Display R2 score

```python
# R2 score
r2_score(y_test,predicted_value)
```
