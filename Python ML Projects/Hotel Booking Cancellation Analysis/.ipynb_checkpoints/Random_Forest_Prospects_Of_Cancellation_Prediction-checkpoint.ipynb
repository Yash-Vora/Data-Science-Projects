# Working with model

Create features and target data

```python
x = df.drop(['is_canceled'],axis=1)
x.head()
```

```python
y = df[['is_canceled']]
y.head()
```

Split into training & testing

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

Apply Random forest classifier on data

```python
RF = RandomForestClassifier()
```

```python
import warnings
warnings.filterwarnings('ignore') # Ignore warnings
RF.fit(x_train,y_train)
```

```python
RF.score(x_train,y_train)
```

```python
RF.score(x_test,y_test)
```

Create function which show Precision score, recall score, accuracy, classification report and confusion matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report
```

```python
def model_score(xt):
    y_pred = RF.predict(xt)
    print("Accuracy Score:\n",accuracy_score(y_test,y_pred),"\n",sep="")
    print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred),"\n",sep="")
    print("Precision Score:\n",precision_score(y_test,y_pred),"\n",sep="")
    print("Recall Score:\n",recall_score(y_test,y_pred),"\n",sep="")
    print("F1 Score:\n",f1_score(y_test,y_pred),"\n",sep="")
    print("Classification Report:\n",classification_report(y_test,y_pred),sep="")
```

```python
model_score(x_test)
```
